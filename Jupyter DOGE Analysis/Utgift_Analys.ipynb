```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("DOGE DEMO.csv")
df.head()
```

```python
mynd = "Naturvårdsverket"
year = 2024
filter_df = df[(df['Myndighet'] == mynd) & (df['År'] == year)]
filter_df

total_summa = filter_df['Total'].sum().sum()
print("Total Money (B):", total_summa/1000000000)
```

```python
filter_df = df[df['År'] == 2022]

grupp_df = filter_df.groupby('Myndighet')['Total'].sum()
sum_total_22 = grupp_df.sum()
print(sum_total_22/1000000000)
grupp_df.sort_values(ascending=False)
```
